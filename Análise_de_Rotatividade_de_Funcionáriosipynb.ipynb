# Interpretação

* Acurácia: Representa a proporção de previsões corretas que o modelo fez em relação ao total de previsões.

* Precision: Indica a proporção de verdadeiros positivos (funcionários que deixaram a empresa previstos corretamente) entre todos os casos previstos como positivos pelo modelo.

* Recall (Sensibilidade): Refere-se à proporção de verdadeiros positivos identificados corretamente pelo modelo entre todos os casos positivos reais.

* F1-score: É a média harmônica entre precision e recall, útil quando se deseja uma métrica balanceada entre precision e recall.

* AUC ROC: Representa a capacidade do modelo de distinguir entre classes positivas e negativas. Quanto maior o valor, melhor o desempenho do modelo em classificar as observações.
